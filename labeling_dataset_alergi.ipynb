```python
import pandas as pd
```

```python
dataset = pd.read_csv('dataset alergi 2.csv')
```

```python
dataset
```

```python
#merubah nilai dengan label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
#menimpa dataset riwaya_alergi dengan encoder
dataset['riwayat_alergi'] = encoder.fit_transform(dataset['riwayat_alergi'])
```

```python
dataset
```

```python
for column in dataset.columns:
    dataset[column] = encoder.fit_transform(dataset[column])
```

```python
dataset
```

```python
#menyimpan file dataset yang sudah dirubah
dataset.to_csv('dataset_labeling.csv', index=False)
```
